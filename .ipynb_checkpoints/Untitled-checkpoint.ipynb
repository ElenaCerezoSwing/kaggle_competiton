```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./data/cars_train.csv')
train.head()
```

```python
train.columns
```

```python
def cylinders_to_string(s):
    if isinstance(s, str) and 'cylinders' in s:
        return int(s.split(' ')[0])
    else:
        return np.nan
```

```python
train['cylinders'] = train['cylinders'].apply(cylinders_to_string)
```

train.city.value_counts()

train.state_code.value_counts()

```python
train.drop(['city', 'county_name', 'county_fips',
            'make', 'manufacturer', 'paint_color',
           'state_code', 'state_name', 'drive', 'title_status'], axis=1, inplace=True)
```

```python
train.transmission.value_counts()
```

```python
def get_transmission_values(s):
    if s =='automatic':
        return 1
    elif s =='manual':
        return 0
    else:
        return np.NaN
```

```python
train['transmission'] = train['transmission'].apply(get_transmission_values)
train.head()
```

```python
test = pd.read_csv('cars_test.csv')
test.head()
```

```python
train = train.dropna()
train.dtypes
```

```python
train = pd.get_dummies(train, ['condition', 'fuel', 'size', 'type'])
```

```python
train.shape
```

```python
y = train['price']
y.head()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(max_iter=1000, tol=1e-3)
sgd.fit(X, y)
```
